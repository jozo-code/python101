삼각함수 그리기
- np.arange(a,b,s): a부터 b까지 s의 간격
- np.sin(value)

````python
import numpy as np
t=np.arange(0,12,0.01)
y=np.sin(t)
````

````python
plt.figure(figsize=(10,6))
plt.plot(t,np.sin(t))
plt.plot(t,np.cos(t))
plt.show()
````

- 격자무늬 추가
- 그래프제목 추가
- x축, y축 제목 추가
- 주황색, 파란색 선 데이터 의미 구분

````python
def graph():
    plt.figure(figsize=(10,6))
    plt.plot(t,np.sin(t),label="sin")
    plt.plot(t,np.cos(t), label="cos")
    plt.grid(True)
    plt.legend(loc="upper left") #범례
    plt.title("Example")
    plt.xlabel("time")
    plt.ylabel("Amlitude")
    plt.show()
````

````python
graph()
````

## scatter plot

````python
t= np.array(range(0,10))
y=np.array([9,8,7,9,8,3,2,4,3,4])
````

````python
plt.figure(figsize=(10,6))
plt.scatter(t,y)
plt.show()
````

````python
colormap=t
def drawGraph():
    plt.figure(figsize=(10,6))
    plt.scatter(t,y, s=150, c=colormap, marker="<")
    plt.colorbar()
    plt.show()
  
drawGraph()
````
